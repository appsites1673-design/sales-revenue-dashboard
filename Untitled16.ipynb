```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv('customers.csv')

# Display first 5 rows
print("Dataset:")
print(df.head())

# Convert 'AnnualIncome' and 'SpendingScore' to numeric, coercing errors to NaN
df['AnnualIncome'] = pd.to_numeric(df['AnnualIncome'], errors='coerce')
df['SpendingScore'] = pd.to_numeric(df['SpendingScore'], errors='coerce')

# Drop rows with NaN values in 'AnnualIncome' or 'SpendingScore'
df.dropna(subset=['AnnualIncome', 'SpendingScore'], inplace=True)

# Select features from the cleaned DataFrame
X = df[['AnnualIncome', 'SpendingScore']]

# Create K-Means model
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto') # Added n_init='auto' to suppress future warning

# Fit model and predict clusters
df['Cluster'] = kmeans.fit_predict(X)

# Display clustered data
print("\nCustomer Segments:")
print(df)

# Display cluster centers
print("\nCluster Centers:")
print(kmeans.cluster_centers_)

# Plot clusters
plt.figure(figsize=(8,6))

plt.scatter(
    df['AnnualIncome'],
    df['SpendingScore'],
    c=df['Cluster'],
    cmap='rainbow',
    s=100
)

# Plot cluster centers
plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=300,
    c='black',
    marker='X',
    label='Centroids'
)

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation using K-Means Clustering')
plt.legend()
plt.grid(True)

plt.show()
```
