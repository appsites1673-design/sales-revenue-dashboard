```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
# loading dummy data for now
df = pd.read_csv("/content/Sales_data.csv")

# 2. Quick Data Check
print(df.head())
print(df.info())

# 3. Clean Data
df.dropna(inplace=True)   # remove missing values
df['Date'] = pd.to_datetime(df['Date'])  # ensure Date column is datetime

# 4. KPIs
total_sales = df['Sales'].sum()
total_revenue = df['Revenue'].sum()
print("Total Sales:", total_sales)
print("Total Revenue:", total_revenue)

# 5. Revenue Trend (Line Chart)
plt.figure(figsize=(10,5))
df.groupby(df['Date'].dt.to_period("M"))['Revenue'].sum().plot(kind='line')
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# 6. Top Products (Bar Chart)
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(5)
sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 5 Products by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Product")
plt.show()

# 7. Sales Distribution (Pie Chart)
sales_by_product = df.groupby('Product')['Sales'].sum()
sales_by_product.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Sales Distribution by Product")
plt.ylabel("")
plt.show()
```
